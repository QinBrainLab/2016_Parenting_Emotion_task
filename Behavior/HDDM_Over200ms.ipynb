```python
import hddm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Load data from csv file into a NumPy structured array
data = hddm.load_csv('E:\\hDDM_ANT_PD-master\\hDDM_ANT_PD-master\\Input.csv')
#data = hddm.load_csv('bd_wm_hddm_byload_2B.csv')
data=data[data['rt']>0.2]
data=data[data['group']!=40]
data=data[data['group']!=50]
```

```python
model = hddm.HDDM(data,p_outlier=0.05,depends_on={'v':'group','a':'group','t':'group'},is_group_model=True)
model.find_starting_values()
# Create model and start MCMC sampling
model.sample(12000, burn=2000,dbname='traces_parent12000.db',db='pickle')#10000,5000
model.save('mymodel_all__12000_2000Burnin')
#print(model.dic)
print('Test is done')
model_stat= model.gen_stats()
model_stat.to_csv(r'mymodel_all_12000_2000Burnin.csv')
model.plot_posterior_predictive()
model.print_stats()
```
